## Univariate Analysis - Only 1 Numeric column

## Dist Plot
- seaborn.distplot() is a function used to visualize the distribution of a dataset by displaying a histogram combined with a kernel density estimate (KDE). It helps to understand the frequency and density of data points in a dataset.

```python
sb.distplot(df['hsc_p'], kde= True, kde_kws={'color':'red'})
```

## Bivariate Vaisualization : Distribution Plot - Only Numerical columns

## Joint Plot
- Seaborn.jointplot() is a function used to visualize the relationship between two variables along with their individual distributions.
- It creates a scatter plot (or other plot types like hex, reg, or kde) to show the relationship, while also displaying marginal histograms or density plots along the x and y axes to represent the distributions of each variable.
-  This provides a comprehensive view of both correlation and distribution in one visual.

```python
sb.jointplot(x='hsc_p',y='salary',data = df)
plt.show()
```

## Joint Plot - heatmap

```python
sb.jointplot(x='hsc_p',y='salary',data = df, kind='hex') 
plt.show()
```

## Joint Plot - Kernal density estimation curve

```python
sb.jointplot(x='hsc_p', y='salary', data=df, kind='kde') 
plt.show()
```

## Pair Plot - Numeric column comparison
- seaborn.pairplot() is a function used to create a grid of scatter plots to visualize the relationships between multiple variables in a dataset.
- It displays scatter plots for every pair of variables, along with histograms or kernel density estimates on the diagonal to show the distribution of each variable.
- This allows for a comprehensive exploration of relationships and correlations in multivariate data, making it useful for identifying patterns and trends.

```python
sb.pairplot(df, hue='gender', kind='scatter',diag_kind='kde', palette = "husl")
```

## Bivariate Visualization - 1 Categorical Column & 1 Numerical column 

## Strip Plot 
- seaborn.stripplot() is a function used to display individual data points along a categorical axis.
- It creates a scatter plot where data points are plotted as dots, which helps visualize the distribution and frequency of data within each category.
- The dots can be jittered (randomly spread) to reduce overlap and improve visibility.
- It is often used in conjunction with other plots, like box plots or violin plots, to provide additional insights into the underlying data distribution within categories

```python
sb.stripplot(x='gender', y = 'hsc_p', data=df)
```

```python
sb.stripplot(x='gender', y = 'hsc_p', data=df, jitter = True)
```

```python
sb.stripplot(x='hsc_b', y = 'hsc_p', data=df)
```

## Swarm Plot
- seaborn.swarmplot() is a function used to visualize the distribution of data points across categories while avoiding overlap.
- It displays individual data points as dots, arranged along a categorical axis, with a slight horizontal jitter to ensure clarity.
- This allows for an accurate representation of the data distribution within each category, making it useful for identifying patterns, clusters, and outliers in the dataset.
- It can be combined with other plots, such as box plots or violin plots, for enhanced visualization.tion. 

```python
sb.swarmplot(x='hsc_b', y = 'hsc_p', data=df)
```

```python
sb.swarmplot(x='hsc_b', y = 'hsc_p', data=df, hue='gender')
```

## Distribution Observation - 1 Categorical & 1 Numerical column

## Violin Plot
- seaborn.violinplot() is a function used to visualize the distribution of a dataset across different categories.
- It combines aspects of a box plot and a kernel density plot, displaying the probability density of the data at different values.
- The width of the "violin" indicates the density of data points, while the center typically shows the median or interquartile range.
- This plot provides a comprehensive view of the distribution, allowing for easy comparison between categories and highlighting variations in data.

```python
sb.violinplot(data=df,x='hsc_b', y='hsc_p',palette = 'husl')
```

```python
sb.violinplot(data=df,x='hsc_b', y='hsc_p',hue='gender')
```

```python
sb.violinplot(x = 'specialisation',y = 'salary',data = df, hue = 'gender')
plt.show()
```

## Boxplot
- seaborn.boxplot() is a function used to visualize the distribution of a dataset through its quartiles.
- It displays a box that represents the interquartile range (IQR), with lines (whiskers) extending to show the range of the data outside the IQR.
- The boxplot also includes markers for the median, outliers, and potential extremes.
- This plot provides a clear summary of the central tendency, variability, and skewness of the data, making it useful for comparing distributions across different categories.

```python
sb.boxplot(x = 'ssc_b', y = 'ssc_p', data = df, hue = 'gender')
plt.show()
```

## Boxplot - For whole data

```python
sb.boxplot(data = df, orient = 'h')
plt.show()
```

## Barplot
- seaborn.barplot() is used to show the average value of a numerical variable for different categories.
- Each bar represents the mean (or other aggregate) for a category, and optional error bars show variability like standard deviation.
- It's useful for comparing averages between groups.

```python
sb.barplot(x = 'ssc_b', y = 'ssc_p', data = df, hue = 'gender')
plt.show()
```

## Countplot
- seaborn.countplot() is a function used to display the count of observations in each category of a categorical variable.
- It shows the frequency of each category as bars, making it easy to compare how many times each category appears in the dataset.
- This plot is useful for visualizing categorical data distribution.

```python
sb.countplot(x = 'ssc_b', data = df, hue = 'gender')
plt.show()
```

## Pointplot
- seaborn.pointplot() is a function used to show the relationship between two categorical variables and one numerical variable.
- It plots points representing the mean of the numerical variable for each category, with lines connecting the points to show trends.
- Error bars can be added to represent variability, such as standard deviation or confidence intervals.
- This plot is useful for highlighting changes or differences between categories.

```python
sb.pointplot(x='ssc_p', y = 'ssc_b', data = df, hue = 'gender')
plt.show()
```

## Regplot
- seaborn.regplot() is a function used to plot data and fit a linear regression model on the data.
- It shows both a scatter plot of the data points and a regression line, which represents the best-fit line through the data.
- The function also includes an option to add a confidence interval around the regression line to visualize the uncertainty of the model.
- It is useful for analyzing the relationship between two numerical variables and determining how well one variable can predict the other.

```python
sb.regplot(x='total_bill', y = 'tip', data = df2)
```

## Lmplot
- seaborn.lmplot() is a function used to plot linear regression models across different subsets of data.
- It combines a scatter plot with a linear regression line, and allows you to easily create multiple regression plots for different categories by using the hue, col, or row parameters.
- This makes it useful for exploring the relationships between variables across various groupings or conditions.
- It provides more flexibility than regplot() for handling multiple categories and subplots.



```python
sb.lmplot(x='total_bill', y = 'tip', col = 'smoker',data = df2)
```

## Lmplot
- This filters the DataFrame df3 to only include rows where the value in the dataset column is 'II'. The query() function is used to subset the DataFrame based on this condition.

```python
sb.lmplot(x = "x", y = "y", data = df3.query("dataset == 'II'"))
```

- order = 2
- ---> Instead of fitting a straight line (first-order linear regression), the function will fit a quadratic curve to the data.
- The regression model will include both x and x^2 terms to capture more complex, nonlinear relationships between x and y.

```python
sb.lmplot(x = "x", y = "y", data = df3.query("dataset == 'II'"), order = 2)
```

## FacetGrid
- seaborn.FacetGrid is a powerful tool that allows you to create a grid of plots based on subsets of your dataset.
- It enables the visualization of data across multiple variables by "faceting" the data into different categories, creating a separate plot for each subset.
- You can specify variables for rows, columns, and different hues (colors) to create complex visualizations that show patterns or trends across multiple dimensions.
- It is useful for exploring data relationships across categorical or continuous variables in a structured way.

```python
grid = sb.FacetGrid(df4, col = "time", row = 'sex', hue = 'sex')
grid.map(sb.histplot, "tip")
grid.add_legend()
plt.show()
```

## PairGrid
- Seaborn's PairGrid is a powerful tool for visualizing relationships between multiple variables in a dataset.
- It creates a grid of subplots for each pair of variables, allowing you to explore their interactions in a systematic way.
- You can customize the plots in each cell of the grid, such as using scatter plots, histograms, or KDE plots, to analyze the relationships and distributions of the data.
- PairGrid is especially useful for examining complex datasets with multiple dimensions and understanding how different features relate to one another.

```python
grid = sb.PairGrid(df5)
grid.map(sb.scatterplot)
grid.map_diag(sb.histplot)
plt.show()
```

```python
grid = sb.PairGrid(df5)
grid.map_offdiag(sb.scatterplot)
grid.map_diag(sb.histplot)
plt.show()
```

```python
grid = sb.PairGrid(df5)
grid.map_offdiag(sb.scatterplot)
grid.map_diag(sb.kdeplot)
grid.map_lower(sb.kdeplot)
plt.show()
```
